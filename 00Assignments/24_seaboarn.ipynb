### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Ans ->
 Seaborn is a powerful Python data visualization library. Here are five types of plots you can create using Seaborn and their uses:

1. **Scatter Plot:**
   - Use: Visualize the relationship between two continuous variables.
   - Example Use Case: Showing the correlation between height and weight.

2. **Bar Plot:**
   - Use: Compare categorical data and display their frequencies or values.
   - Example Use Case: Comparing the sales of different products in a store.

3. **Histogram:**
   - Use: Display the distribution of a single numerical variable.
   - Example Use Case: Visualizing the distribution of ages in a population.

4. **Box Plot (Box-and-Whisker Plot):**
   - Use: Show the distribution of data along a single axis and identify outliers.
   - Example Use Case: Analyzing the spread of exam scores across different subjects.

5. **Heatmap:**
   - Use: Display the correlation matrix between multiple variables.
   - Example Use Case: Understanding the relationships between various features in a dataset.



### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

Note:  pclass, age, and fare are columns in the titanic dataset.

```python
# import seaborn as sns
# titanic = sns.load_dataset('titanic')
# sns.boxplot(data=titanic, x='pclass', y='age', color='blue')
# sns.boxplot(data=titanic, x='pclass', y='fare', color='red')
# sns.show()
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create the first box plot: pclass vs age
plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.show()

# Create the second box plot: pclass vs fare
plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue 
parameter for the 'cut' column of the diamonds dataset. 

```python
import seaborn as sns
data =sns.load_dataset("diamonds")
# sns.histplot(data.price, hue = 'data.cut')
sns.histplot(data=data, x='price', hue='cut', multiple='stack')
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# import seaborn as sns
# data =sns.load_dataset("flights")
# sns.heatmap(data=data)

import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Reshape the data into a pivot table
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Flights Dataset Heatmap')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
